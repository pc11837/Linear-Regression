```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a=pd.read_csv('placement.csv')
```

```python
a.head()
```

```python
plt.scatter(a['TV'],a['Sales'])
plt.xlabel('TV')
plt.ylabel('Sales')
```

```python
x=a.iloc[:,0:1]
y=a.iloc[:,1]
```

```python
x,y
```

```python
from sklearn.model_selection import train_test_split
# now splitting train test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.coef_
lr.intercept_#b that is y intercept
y_pred=lr.predict(x_test)
y_pred
x_test
```

```python
m=lr.coef_
b=lr.intercept_
y=(m*8.58)+b
y
```

```python
plt.scatter(a['TV'],a['Sales'])
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('TV')
plt.ylabel('Sales')
plt.show()
```
